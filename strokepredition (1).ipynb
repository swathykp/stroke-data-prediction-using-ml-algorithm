```python
dummy=pd.get_dummies(df2[['gender','ever_married','work_type','Residence_type','smoking_status']])
dummy
```

```python
#Normalization and Modelling
```

```python
dfe=pd.concat([df2,dummy],axis=1)
dfe
```

```python
dfe=dfe.drop(['gender','ever_married','work_type','Residence_type','smoking_status'],axis=1)
```

```python
dfe
```

```python
# Split data into X and Y
X = dfe.iloc[:,:-1]
Y = dfe['stroke']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
#Data normalization using scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn import model_selection
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

classifier_KNN = KNeighborsClassifier()
classifier_DT = DecisionTreeClassifier()
classifier_RF = RandomForestClassifier()
```

```python
classifier_KNN.fit(X_train, y_train)
```

```python
y_pred = classifier_KNN.predict(X_test)
dataEvaluation=pd.DataFrame({"y_Pred":y_pred, "y_Actual":y_test})
dataEvaluation
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score,ConfusionMatrixDisplay
result1=confusion_matrix(y_test,y_pred)
print("confusion matrix:")
print(result1)
result2=accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
```

```python
classifier_DT.fit(X_train, y_train)
```

```python
y_pred1 = classifier_DT.predict(X_test)
dataEvaluation=pd.DataFrame({"y_Pred1":y_pred1, "y_Actual":y_test})
dataEvaluation
```

```python
result_dt=confusion_matrix(y_test,y_pred1)
print("confusion matrix:")
print(result_dt)
result_dt1=accuracy_score(y_test,y_pred1)
print("Accuracy:",result_dt1)
```

```python
classifier_RF.fit(X_train, y_train)
```

```python
y_pred2 = classifier_RF.predict(X_test)
dataEvaluation=pd.DataFrame({"y_Pred2":y_pred2, "y_Actual":y_test})
dataEvaluation
```

```python
result_rf=confusion_matrix(y_test,y_pred2)
print("confusion matrix:")
print(result_rf)
result_rf1=accuracy_score(y_test,y_pred2)
print("Accuracy:",result_rf1)
```
